```python
#简化模型
import numpy as np
net = np.array([[0,0,0,1/3,0,0],
              [1/4,0,0,0,1/2,0],
              [0,1,0,1/3,1/2,0],
              [1/4,0,0,0,0,1],
              [1/4,0,1,1/3,0,0],
              [1/4,0,0,0,0,0]])
b = np.array([1/6,1/6,1/6,1/6,1/6,1/6])
def simple(x,y):
    for i in range(100):
        y=np.dot(x,y)
    return y
result = simple(net,b)
print(result)
```

```python
#随机模型
def random(x,y,n):
    d=0.85
    for i in range(100):
        y=(1-d)/n+d*np.dot(x,y)
        print(y)
random(net,b,6)
```
